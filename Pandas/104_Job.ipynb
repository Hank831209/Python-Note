```python
import pandas as pd
import requests
area_json = 'https://static.104.com.tw/category-tool/json/Area.json'
response = requests.get(url=area_json, headers=headers)
response.json()[0]['n']
df = pd.DataFrame(response.json()[0]['n'])
# df
df = df.explode('n')
a = df['n'].apply(lambda x: x['des'])
df['des2'] = df['n'].apply(lambda x: x['des'])
df['no2'] = df['n'].apply(lambda x: x['no'])
df_area = df.loc[:, ['des', 'no', 'des2', 'no2']]
df_area
df_area['no2'].unique()
df_area
```

```python
import pandas as pd
import requests
import os
import time
start_time = time.time()
ro = '1'  # 0: 全部, 1: 全職, 2: 兼職
isnew = '7'  # 幾日內更新的工作
keyword = '軟體工程師'
df = pd.DataFrame(columns=search_list.keys())
df = list()
for area in df_area['no2'].unique():
    for page in range(1, 2):
        url = f'https://www.104.com.tw/jobs/search/list?ro={ro}&isnew={isnew}&kwop=7&keyword={keyword}&area={area}&order=15&asc=0&page={page}&mode=l'
        response = requests.get(url=url, headers=headers)
        ndf = pd.DataFrame(response.json()['data']['list'])
        df.append(ndf)
df = pd.concat(ndf, ignore_index=True)
df

if not os.path.exists(folder_path):
    os.mkdir(folder_path)
df.to_csv(folder_path + f'/{keyword}_job_search.csv', index=0, encoding='utf-8-sig')
end_time = time.time()
print('運算時間:\t', end_time - start_time)
df_area['no2'].unique()

        
```
