# 數據可視化
## 散點圖: scatter

```python
import numpy as np
import pandas as pd

n = 1024    # data size
df = pd.DataFrame({
    "x": np.random.normal(0, 1, n),
    "y": np.random.normal(0, 1, n), 
})
print('df:\n', df)
color = np.arctan2(df["y"], df["x"])
# c: 顏色, s: 點的大小, alpha: 透明度, cmap: colormap
df.plot.scatter(x="x", y="y", c=color, s=60, alpha=0.5, cmap="rainbow")
```

## 折線圖: plot

```python
import numpy as np
import pandas as pd

n = 20    # data size
x = np.linspace(-1, 1, n)
y = x * 2 + 0.4 + np.random.normal(0, 0.3, n)
df = pd.DataFrame({"x": x, "y": y})
# print('df:\n', df)
df.plot(x="x", y="y", alpha=0.5, c="r")
```

```python
import numpy as np
import pandas as pd

n = 20    # data size
x = np.linspace(-1, 1, n)
y1 = x * -1 - 0.1 + np.random.normal(0, 0.3, n)
y2 = x * 2 + 0.4 + np.random.normal(0, 0.3, n)
df = pd.DataFrame({"x": x, "y1": y1, "y2": y2})
df.plot(x="x", y=["y1", "y2"], alpha=0.5)
```

## 長條圖: bar

```python
import numpy as np
import pandas as pd

df = pd.DataFrame(np.random.rand(5, 3), columns=["a", "b", "c"])
df.plot.bar()
```

```python
import numpy as np
import pandas as pd

df = pd.DataFrame(np.random.rand(5, 3), columns=["a", "b", "c"])
df.plot.bar(stacked=True)  # 畫一起
```

```python
import numpy as np
import pandas as pd

df = pd.DataFrame(np.random.rand(5, 3), columns=["a", "b", "c"])
df.plot.barh()  # 橫著畫, h: horizontal
```

## 直方圖: hist(histogram)

```python
import numpy as np
import pandas as pd

df = pd.DataFrame({"a": np.random.randn(1000)})
# print(df)
df.plot.hist()
```

```python
import numpy as np
import pandas as pd

df = pd.DataFrame(
    {
        "a": np.random.randn(1000) + 1,
        "b": np.random.randn(1000),
        "c": np.random.randn(1000) - 4,
    }
)

df.plot.hist(alpha=0.5, bins=30)  # bins: 柱體的數量
```

## 圓餅圖: pie

```python
import numpy as np
import pandas as pd

df = pd.DataFrame(
    {"boss": np.random.rand(4)},
    index=["meeting", "supervise", "teaching", "team building"], 
)
df.plot.pie(y="boss", figsize=(7,7))
```

```python
import numpy as np
import pandas as pd

df = pd.DataFrame(
    {
        "bigBoss": np.random.rand(4),
        "smallBoss": np.random.rand(4),
    },
    index=["meeting", "supervise", "teaching", "team building"], 
)
df.plot.pie(subplots=True, figsize=(9,9), legend=False)
```
